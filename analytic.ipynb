```python

error_list = {'code': [], 'address': []}
code_list = []
address_list = []
location = []
```

```python
# Lấy tọa độ của các điểm trong dataset: 
import time


for i in range(355, len(data_buucuc)):
    if type(data_buucuc.iloc[i]['Address']) is float: found_flag = False
    else:
        old_address = data_buucuc.iloc[i]['Address'].split('´')
        if len(old_address) == 1: old_address = old_address[0].split(' ')
        elif len(old_address) == 2: old_address = old_address[1].split(' ')
        found_flag = False
        for j in range(len(old_address)):
            try:
                getLoc = loc.geocode(' '.join(old_address[j:]) +', ' + str(data_buucuc.iloc[i]['CommuneCode']))
            except: break
            # time.sleep(0.5)
            if getLoc is not None: 
                code_list.append(data_buucuc.iloc[i]['POSCode'])
                address_list.append(getLoc.address)
                location.append([getLoc.latitude, getLoc.longitude])
                found_flag = True
                break
    
    if i%100 == 0: 
        print(f'Done{i}/{len(data_buucuc)}')
    if not found_flag: 
        error_list['code'].append(data_buucuc.iloc[i]['POSCode'])
        error_list['address'].append(data_buucuc.iloc[i]['Address'])
            
    
```

```python
import pickle as pkl
pkl.dump({'code': code_list, 'address': address_list, 'location': location, 'error': error_list}, open('data/vnpost.pkl', 'wb'))
```
